```python
import pandas as pd
import seaborn as sns
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
sinasc_ro = pd.read_csv('SINASC_RO_2019.csv')
```

```python
# mostrando os nomes das colunas/atributos
sinasc_ro.columns
```

```python
sns.boxplot(data=sinasc_ro, y='IDADEMAE')
```

####  2) Faça um boxplot do peso do bebê

```python
sinasc_ro[['PESO']].boxplot()
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(data=sinasc_ro, x='IDADEMAE', alpha=0.4, kde=True, bins=15, element='step', color='g')
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(data=sinasc_ro, x='PESO', kde=True, bins=20, element='step', color='magenta', alpha=0.25)
```
