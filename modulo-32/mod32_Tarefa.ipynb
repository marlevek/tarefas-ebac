# Probabilidade - Tarefa

```python
import numpy as np
from scipy import stats
```

1. Uma mquina embala rao para cachorro em pacotes de 15 kg, mas cada pacote produzido acaba tendo um peso aleatrio, com distribuio Normal, com mdia 15kg, mas com desvio padro 0,1 kg. Obtenha a probabilidade de que um pacote selecionado aleatoriamente dessa produo tenha mais que 15,2 kg.

```python
media = 15
desvio_padrao = 0.1

# probabilidade acumulada at 15,2 kg
prob_ate_15_2 = stats.norm.cdf(15.2, loc=media, scale=desvio_padrao)

# A probabilidade de que um pacote tenha mais que 15,2 kg
prob_maior_15_2 = 1 - prob_ate_15_2

print(f'A probabilidade de que um pacote tenha mais de 15,2 kg  {prob_maior_15_2:.4f}')
```

2. Da mquina descrita acima, obtenha o percentil de ordem 95% para o peso dos pacotes. Ou seja, o valor tal que a probabilidade de um pacote ter peso menor que este seja de 95%.

```python
# Calcular o percentil de ordem 95%
percentil_95 = stats.norm.ppf(0.95, loc=media, scale=desvio_padrao)

print(f'O percentil de ordem 95% para o peso dos pacotes  {percentil_95:.4f} kg')
```

3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, a probabilidade de ser menor que cada um dos valores contidos no objeto $Y$ na clula abaixo:

```python
Y=np.linspace(-6, 6, 200)

# Seu cdigo comea aqui
Y = [-2.0, -1.0, 0.0, 1.0, 2.0]

# Probabilidades para a varivel Y1 (Normal Padro)
probs_Y1 = [stats.norm.cdf(y) for y in Y]

# Probabilidades para a varivel Y2 (t de Student com 5 graus de liberdade)
probs_Y2 = [stats.t.cdf(y, df=5) for y in Y]

print("Probabilidades para a varivel Y1 (Normal Padro):", probs_Y1)
print("Probabilidades para a varivel Y2 (t de Student com 5 graus de liberdade):", probs_Y2)
```

4. Faa uma visualizao grfica dos dados do item anterior.

```python
import matplotlib.pyplot as plt 

# Probabilidades calculadas anteriormente
probs_Y1 = [0.022750131948179195, 0.15865525393145707, 0.5, 0.8413447460685429, 0.9772498680518208]
probs_Y2 = [0.05096973941492914, 0.18160873382456127, 0.5, 0.8183912661754387, 0.9490302605850709]

# Criar o grfico
plt.figure(figsize=(10, 6))
plt.plot(Y, probs_Y1, marker='o', label='Normal Padro (Y1)')
plt.plot(Y, probs_Y2, marker='x', label='t de Student (5 df) (Y2)')

# Adicionar ttulo e rtulos
plt.title('Probabilidades de Y1 e Y2')
plt.xlabel('Valores de Y')
plt.ylabel('Probabilidade P(Y < y)')
plt.legend()

# Mostrar o grfico
plt.grid(True)
plt.show()
```

5. 3. Suponha duas variveis aleatrias $Y_1$ e $Y_2$, a primeira com distribuio Normal Padro (mdia=0, varincia=1), a segunda com distribuio *t-studdent* com 5 graus de liberdade (tambm padro). Calcule, para cada uma, o percentil (a inversa da *CDF*) para cada um dos valores do objeto *P* abaixo:

```python
Y=np.linspace(0, 1, 20)

# Seu cdigo comea aqui

# Valores para os quais queremos calcular os percentis
P = [0.01, 0.05, 0.5, 0.95, 0.99]

# Percentis para a varivel Y1 (Normal Padro)
percentis_Y1 = [stats.norm.ppf(p) for p in P]

# Percentis para a varivel Y2 (t de Student com 5 graus de liberdade)
percentis_Y2 = [stats.t.ppf(p, df=5) for p in P]

print("Percentis para a varivel Y1 (Normal Padro):", percentis_Y1)
print("Percentis para a varivel Y2 (t de Student com 5 graus de liberdade):", percentis_Y2)
```

6. Faa um grfico com os dados do item anterior.

```python
# Percentis calculados anteriormente
percentis_Y1 = [stats.norm.ppf(p) for p in P]
percentis_Y2 = [stats.t.ppf(p, df=5) for p in P]

# Criar o grfico
plt.figure(figsize=(10, 6))
plt.plot(P, percentis_Y1, marker='o', label='Normal Padro (Y1)')
plt.plot(P, percentis_Y2, marker='x', label='t de Student (5 df) (Y2)')

# Adicionar ttulo e rtulos
plt.title('Percentis de Y1 e Y2')
plt.xlabel('Probabilidades P')
plt.ylabel('Percentil correspondente')
plt.legend()

# Mostrar o grfico
plt.grid(True)
plt.show()
```
