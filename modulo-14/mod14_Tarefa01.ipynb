```python
# Código para gerar os gráficos
def plota_pivot_table(arquivo, value, index, func, ylabel, xlabel, opcao='nada'):
    df = pd.read_csv('SINASC_RO_2019_ABR.csv')
    pivot_table = pd.pivot_table(df, values=value, index=index, aggfunc=func)
    
    if opcao == 'sort':
        pivot_table = pivot_table.sort_values(by=value)
    elif opcao == 'unstack':
        pivot_table = pivot_table.unstack()
    
    pivot_table.plot(figsize=[15, 5])
    
```

```python
base_dados = 'SINASC_RO_2019_MAR.csv'
plota_pivot_table(base_dados, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento' )
```

```python
base_dados = 'SINASC_RO_2019_ABR.csv'
plota_pivot_table(base_dados, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento' )
```

```python
base_dados = 'SINASC_RO_2019_MAI.cs'
plota_pivot_table(base_dados, value='APGAR1', index='DTNASC', func='mean', ylabel='Valor Médio', xlabel='Mês', opcao='sort')
```

```python
base_dados = 'SINASC_RO_2019_JUN.cs'
plota_pivot_table(base_dados, value='APGAR5', index='DTNASC', func='count', ylabel='Valor Médio', xlabel='Mês', opcao='sort')
```

```python
base_dados = 'SINASC_RO_2019_DEZ.csv'
plota_pivot_table(base_dados, value='GRAVIDEZ', index='ESCMAE', func='count', ylabel='Valor Médio', xlabel='Mês', opcao='sort')
```
