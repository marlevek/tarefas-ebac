#### 2. Observando a volatilidade dentro do dia
1. Calcule a amplitude de variação do valor do ativo dentro do dia. Essa amplitude você pode calcular com um comando simples de adicionar uma coluna no Pandas: (valor máximo - valor mínimo).
2. Monte um gráfico de linha mostrando a variação dessa amplitude ao longo do período coletado, por ativo.
3. Veja se consegue identificar que ativo apresentou maior volatilidade.

```python
import yfinance as yf

# Definir o ativo e a data de início (ativo Tesla)
ativo = 'TSLA'
data_inicio = '2022-01-01'

# Obter os dados do Yahoo Finance até a data atual
ativo_tesla = yf.download(ativo, start=data_inicio)

# Mostrar as primeiras linhas dos dados
print(ativo_tesla.head())
```

```python
# Calcular a amplitude de variação diária (High - Low)
ativo_tesla['Amplitude'] = ativo_tesla['High'] - ativo_tesla['Low']

# Mostrar as primeiras linhas dos dados com a nova coluna
print(ativo_tesla.head())

# Salvar os dados em um arquivo CSV
arquivo_csv = 'tesla_data_with_amplitude.csv'
ativo_tesla.to_csv(arquivo_csv)
```

```python
# Criar o gráfico de linha usando plotly.express
fig = px.line(ativo_tesla, x=ativo_tesla.index, y='Amplitude', title='Variação da Amplitude Diária do Ativo TSLA',
              labels={'Amplitude': 'Amplitude (High - Low)', 'index': 'Data'})

# Mostrar o gráfico
fig.show()
```
